```python
profession=d1["Profession"].value_counts().head(10).values
name=d1["Profession"].value_counts().head(10).index
plt.gcf().set_size_inches(30,15)
plt.pie(profession,labels=name,
        ,wedgeprops={'edgecolor':"k"},explode=[0,0,0,0,0,0,0,0,0.2])
plt.title("Top 10")
plt.show()
```
